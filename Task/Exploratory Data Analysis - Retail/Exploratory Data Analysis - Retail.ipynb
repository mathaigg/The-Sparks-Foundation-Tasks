## profit vs discount

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

# Interpretations
No correlation between profit and discount

## Profit vs Quantity

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

## Interpretations
Profit and sales are maximum in consumer segment and minimum in Home Office segment

## Segment-wise Sales

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index)
plt.xlabel('Region')
plt.show()
```

## Interpretations
Segment wise sales are almost same in every region

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

## Interpretations
Profit and sales are 
- Maximum in west region 
- Minimum in south region

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['green','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states', size=20)
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

## Interpretations
- High profit: California, New York
- High loss: Texas, pennsylvania, Ohio, Illinois

```python
states = df['State'].value_counts().nlargest(10)
states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

## Interpretations
- Technology and Office Supplies have high profit.
- Furniture have less profit & should be looked into to increase its profit

## Profit/Loss vs Sales in each Reagion

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightgreen'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

## Interpretations
- Phones sub-category have high sales.
- Chairs have high sales but less profit compared to phones
- Tables and Bookmarks sub-categories are facing huge loss
